```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.to_csv("Entregable1.csv")
```

```python
#Filtra los registros de la columna NOMBRE_VENDEDOR que tengan el nombre LETICIA RAMIREZ HERNANDEZ
filtro1=df[(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro1
```

```python
filtro1.to_csv("Entregable_filtro1.csv")
```

```python
#Filtra las filas de la 2 a la 9
filtro2= df.iloc[3:10,: ]
filtro2
```

```python
filtro2.to_csv("Entregable_filtro2.csv")
```

```python
#Filtra las columnas 2, 5 y 6
filtro3= df.iloc[ : , [1,4,5]]
filtro3
```

```python
filtro3.to_csv("Entregable_filtro3.csv")
```

```python
#Muestra a FECHA_DOC como el indice de la base de datos
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=1)
df1
```

```python
df1.to_csv("Entregable_df1.csv")
```

```python
#Muestra todos los registros de las columnas NOMBRE_VENDEDOR, NOMBRE_CLIENTE y CANT
filtro4=df1.loc[:,["NOMBRE_VENDEDOR","NOMBRE_CLIENTE","CANT"]]
filtro4
```

```python
filtro4.to_csv("Entregable_filtro4.csv")
```

```python
#Muestra los primeros registros de la base de datos, en este caso los primeros 5
filtro5=df.head()
filtro5
```

```python
filtro5.to_csv("Entregable_filtro5.csv")
```

```python
#Muestra los ultimos registros de la base de datos, en este caso los ultimos 5
filtro6=df.tail()
filtro6
```

```python
filtro6.to_csv("Entregable_filtro6.csv")
```

```python
#Muestra los registros menores a 10000
filtro7=df[df["PRECIO_UNITARIO"] < 10000]
filtro7
```

```python
filtro7.to_csv("Entregable_filtro7.csv")
```

```python
#utiliza el operador logico 'y', muestra todos los registros que PRECIO_UNITARIO sea menor que 10000 Y COSTO_TOTAL_CALCULADO sea igual a 1605.58
filtro8= df[(df["PRECIO_UNITARIO"] < 10000) & (df["COSTO_TOTAL_CALCULADO"] == 1606.58)]
filtro8
```

```python
filtro8.to_csv("Entregable_filtro8.csv")
```

```python
#utiliza el operador logico 'o', muestra todos los registros que PRECIO_UNITARIO sea menor que 10000 Ó COSTO_TOTAL_CALCULADO sea igual a 1605.58
filtro9= df[(df["PRECIO_UNITARIO"] < 10000) | (df["COSTO_TOTAL_CALCULADO"] == 1606.58)]
filtro9
```

```python
filtro9.to_csv("Entregable_filtro9.csv")
```

```python
#muestra todos los registros EXCEPTO los que en PRECIO_UNITARIO sean menores que 10000
filtro10=df[~(df["PRECIO_UNITARIO"] < 10000)]
filtro10
```

```python
filtro10.to_csv("Entregable_filtro10.csv")
```
