```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('listings.xlsx')
df.to_csv("Entregable_df.csv")
df
```

```python
#a)los registros de 5 "host name" diferentes
filtro1=df[(df["name"] == "alberti view") | (df["name"] == "Casa degli Aliotti") | (df["name"] == "Florence Central Double Private Bathroom") | (df["name"] == "Celestial") | (df["name"] == "Relax and charme in Santa Maria Novella")]
filtro1.to_csv("Entregable_filtro1.csv")
filtro1
```

```python
#b)los registros de los host que se hayan unido a Airbnb "antes del 2019"
filtro2=df[df["host_since"] >= "01/01/2019"]
filtro2.to_csv("Entregable_filtro2.csv")
filtro2
```

```python
#c) los registros de los host que responden "a mas tardar en 1 dia"
filtro3=df[df["host_response_time"] == "within a day"]
filtro3.to_csv("Entregable_filtro3.csv")
filtro3
```

```python
#d)Los registros que tengan al menos 5 accommodates
filtro4=df[df["accommodates"] >= 5]
filtro4.to_csv("Entregable_filtro4.csv")
filtro4
```

```python
#e) Los registros de los tipos de cuarto "Entire home" y "Private room"
filtro5=df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5.to_csv("Entregable_filtro5.csv")
filtro5
```

```python
#f)los registros que "a lo mucho cuenten con 4 camas"
filtro6=df[(df["beds"] <= 4)]
filtro6.to_csv("Entregable_filtro6.csv")
filtro6
```

```python
#g) Los registros que "tengan 1 baño o que sean superhost" pero con la condición de que "todos sean instant_bookable"
filtro7=df[((df["bathrooms"] <= 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
filtro7.to_csv("Entregable_filtro7.csv")
filtro7
```

```python
#h)Los registros cuyo "review_record_cleanliness sea menor de 3 y review_score_rating sea mayor de 3"
filtro8=df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8.to_csv("Entregable_filtro8.csv")
filtro8
```
